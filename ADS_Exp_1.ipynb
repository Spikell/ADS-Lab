```python
# Frequency distribution for categorical variables
print("\nFrequency Distribution for Categorical Variables:")
print(df['Sex'].value_counts())
print(df['Diabetes'].value_counts())
print(df['Smoking'].value_counts())
print(df['Family History'].value_counts())
print(df['Country'].value_counts())


# Histograms for numerical variables
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(df['Cholesterol'], bins=20, color='salmon', edgecolor='black')
plt.title('Cholesterol Distribution')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(df['Sedentary Hours Per Day'], bins=20, color='orange', edgecolor='black')
plt.title('Sedentary Hours Per Day Distribution')
plt.xlabel('Sedentary Hours Per Day')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(df['Physical Activity Days Per Week'], bins=20, color='green', edgecolor='black')
plt.title('Physical Activity Days Per Week Distribution')
plt.xlabel('Physical Activity Days Per Week')
plt.ylabel('Frequency')


plt.tight_layout()
plt.show()
```

```python
categorical_features = ['Sex', 'Diabetes', 'Family History', 'Smoking', 'Heart Attack Risk', 'Alcohol Consumption', 'Stress Level', 'Continent', 'Hemisphere']

# Calculate the number of rows and columns for subplots
num_plots = len(categorical_features)
num_cols = 3
num_rows = - (- num_plots // num_cols)  # Ceiling division to ensure enough rows

# Set up the figure with the desired number of subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(12, num_rows * 3))

# Flatten the axes array to facilitate iteration
axes = axes.flatten()

# Iterate through each categorical feature and create a count plot
for i, feature in enumerate(categorical_features):
    sns.countplot(x=feature, data=df, ax=axes[i])
    axes[i].set_title(f'Count Plot of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Count')

# Hide unused subplots if there are any
for j in range(i + 1, num_rows * num_cols):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
